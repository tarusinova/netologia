```python
keywords = pd.read_csv(r'C:\Users\operator\Documents\python\доп_материалы_к_пандас_1_и_2\keywords.csv')
```

```python
geo_data = {
    'Центр': ['москва', 'тула', 'ярославль'],
    'Северо-Запад': ['петербург', 'псков', 'мурманск'],
    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
```

```python
def geo(words: str):
    
    region = []
    
    for key, val in geo_data.items():
        
        for v in val:
            if v in words:
                region.append(key)
    
    if region == []:
        return 'undefined'
    
    elif len(region) == 1:
        return region[0]
    
    else:
        return ', '.join(region)
```

```python
keywords['region'] = keywords['keyword'].apply(geo)
```

```python
movies_ = pd.read_csv(r'C:\Users\operator\Documents\python\доп_материалы_к_пандас_1_и_2\ml-latest-small\movies.csv')
rating_ = pd.read_csv(r'C:\Users\operator\Documents\python\доп_материалы_к_пандас_1_и_2\ml-latest-small\ratings.csv')
```

```python
df = rating_.merge(movies_, how='left', on='movieId')
```

```python
df
```

```python
years = [str(i) for i in range(1950, 2011)]
```

```python
def production_year(name):
    
    year_str = []
    
    for i in years:
        
        if i in name:
            year_str.append(i)
    
    if year_str == []:
        return 1900
    
    elif len(year_str) == 1:
        return int(year_str[0])
    
    else:
        return int(year_str[0])                 #', '.join(year_str)
```

```python
df['year'] = df['title'].apply(production_year)
```
